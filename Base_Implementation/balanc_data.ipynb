```python
from imblearn.over_sampling import SMOTE
import pandas as pd
```

```python
taiwan_CFS=pd.read_csv('../Datasets/CFS/taiwan_CFS.csv')
australia_CFS=pd.read_csv('../Datasets/CFS/australian_CFS.csv')
taiwan=pd.read_csv('../Datasets/Taiwan_credit_approval.csv')
australia=pd.read_csv('../Datasets/Australian_credit_approval.csv')
```

```python
def SMOTE_function(df, target,random_state=404131029):
    X=df.drop(target, axis=1)
    Y=df[target]
    smote=SMOTE(random_state=random_state)
    x_resampled,y_resampled=smote.fit_resample(X,Y)
    df_resampled = pd.DataFrame(x_resampled, columns=X.columns)
    df_resampled[target] = y_resampled
    return df_resampled
```

```python
australia_CFS.columns
```

```python
taiwan_CFS_SMOTE=SMOTE_function(taiwan_CFS,'default.payment.next.month')
taiwan_SMOTE=SMOTE_function(taiwan,'default.payment.next.month')
australia_CFS_SMOTE=SMOTE_function(australia_CFS,'Class')
australia_SMOTE=SMOTE_function(australia,'Class')
```

```python
taiwan_CFS_SMOTE.to_csv("../Datasets/SMOTE/taiwan_CFS_SMOTE.csv")
taiwan_SMOTE.to_csv("../Datasets/SMOTE/taiwan_SMOTE.csv")
australia_CFS_SMOTE.to_csv("../Datasets/SMOTE/australia_CFS_SMOTE.csv")
australia_SMOTE.to_csv("../Datasets/SMOTE/australia_SMOTE.csv")
```
