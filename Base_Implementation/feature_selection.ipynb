```python
Demographic_indicators=["SEX","AGE","EDUCATION","MARRIAGE"]
Finance_indicators=["LIMIT_BAL"]
Behavioral_indicators=["PAY_0","PAY_2","PAY_3","PAY_4","PAY_5","PAY_6","BILL_AMT1","BILL_AMT2","BILL_AMT3","BILL_AMT4","BILL_AMT5","BILL_AMT6","PAY_AMT1","PAY_AMT2","PAY_AMT3","PAY_AMT4","PAY_AMT5","PAY_AMT6"]
```

```python
#T-test


def T_test(feature, target):
    group0 = feature[target == 0]
    group1 = feature[target == 1]
    if len(group0) < 2 or len(group1) < 2:
        return None  
    t_stat, p_val = ttest_ind(group0, group1)
    return round(t_stat, 4)
```

## Feature selection

```python
#R

# Demographic_indicator
print("Demographic_indicator: ")
for column in Demographic_indicators:
    print(f"{column} R={pearsonr(data[column],data['default.payment.next.month'])[0]} \n{column} P_value ={pearsonr(data[column],data['default.payment.next.month'])[1]}")
    print(f"{column} T-test={T_test(data[column],data['default.payment.next.month'])}")
# Finance_indicator
print("Financial_indicator: ")
for column in Finance_indicators:
    print(f"{column} R={pearsonr(data[column],data['default.payment.next.month'])[0]} \n{column} P_value ={pearsonr(data[column],data['default.payment.next.month'])[1]}")
    print(f"{column} T-test={T_test(data[column],data['default.payment.next.month'])}")

#Behavioral_indicator
print("Behavioral_indicator :")

for column in Behavioral_indicators:
  
    print(f"{column} R={pearsonr(data[column],data['default.payment.next.month'])[0]} \n{column} P_value ={pearsonr(data[column],data['default.payment.next.month'])[1]}")
    print(f"{column} T-test={T_test(data[column],data['default.payment.next.month'])}")
```
