## Feature selection

```python
from scipy import stats
```

```python

def correlation_FS(df,src_column,target_column):
    X=df[src_column]
    Y=df[target_column]
    r, p_value = stats.pearsonr(X, Y)
    n = len(X)
    DF = n - 2
    t_value = r * np.sqrt(DF / (1 - r**2))

    print(f"Column {src_column}")
    print("Pearson R:", r)
    print("t-value:", t_value)
    print("p-value:", p_value)
    print("Degrees of Freedom (DF):", DF/n)
    print("\n\n\n")
```

```python
data_taiwan.drop(columns=["ID"],inplace=True)
data_australian.drop(columns=["CustomerID"],inplace=True)
```

```python
for i in data_taiwan.columns:
    correlation_FS(data_taiwan,i,"default.payment.next.month")
```

```python
for i in data_australian.columns:
    correlation_FS(data_australian,i,"Class")
```

```python
def SFS_correlation(df, target_col, alpha=0.01):

    features = [col for col in df.columns if col != target_col]
    selected_features = []

    while True:
        best_feature = None
        best_score = 0
        best_p = 1

        for f in features:
            if f in selected_features:
                continue

            r, p = stats.pearsonr(df[f], df[target_col])
            score = abs(r)

            if p < alpha and score > best_score:
                best_score = score
                best_feature = f
                best_p = p

        if best_feature is None:
            break

        selected_features.append(best_feature)
        print(f"Selected: {best_feature} | |r| = {best_score:.4f} | p = {best_p:.4f}")

    return selected_features
```

```python
selected_features_taiwan_SFS=SFS_correlation(data_taiwan,"default.payment.next.month")
```

```python
selected_features_australian_SFS=SFS_correlation(data_australian,"Class")
```

```python
selected_features_australian_CFS=["A2","A3","A4","A5","A6","A7","A8","A9","A10","A12","A13","A14"]
```

```python
df_taiwan_CFS=data_taiwan.drop(columns=["SEX","EDUCATION","MARRIAGE"]).copy()
df_australian_CFS=data_australian[selected_features_australian_CFS].copy()
df_australian_SFS=data_australian[selected_features_australian_SFS].copy()
df_taiwan_SFS=data_taiwan[selected_features_taiwan_SFS].copy()
```
